```python
import pandas as pd
column_names = [
    "duration","protocol_type","service","flag","src_bytes","dst_bytes",
    "land","wrong_fragment","urgent","hot","num_failed_logins","logged_in",
    "num_compromised","root_shell","su_attempted","num_root","num_file_creations",
    "num_shells","num_access_files","num_outbound_cmds","is_host_login",
    "is_guest_login","count","srv_count","serror_rate","srv_serror_rate",
    "rerror_rate","srv_rerror_rate","same_srv_rate","diff_srv_rate",
    "srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate",
    "dst_host_same_src_port_rate","dst_host_srv_diff_host_rate",
    "dst_host_serror_rate","dst_host_srv_serror_rate","dst_host_rerror_rate",
    "dst_host_srv_rerror_rate","label"
]

# Load the file (adjust path based on your system)
df = pd.read_csv("data/kddcup.data_10_percent.gz",names=column_names)
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['label'] = df['label'].str.strip().str.lower()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of normal vs attack
df['label'].value_counts().plot(kind='bar', figsize=(8,4), color='skyblue')
plt.title('Distribution of Normal vs Attack Traffic')
plt.show()

# Show top 10 attack types (optional)
df['label'].value_counts().head(10)
```

```python
df['target']=df['label'].apply(lambda x : 0 if x == 'normal' else 1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode protocol_type, service, flag
categorical_cols = ['protocol_type', 'service', 'flag']
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.preprocessing import MinMaxScaler

# Drop label columns for now
X = df.drop(['label', 'target'], axis=1)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Create final dataframe
import numpy as np
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df['target'] = df['target']  # Add target back
```

```python
X_scaled_df.to_csv('data/cleaned_kdd.csv', index=False)
```
